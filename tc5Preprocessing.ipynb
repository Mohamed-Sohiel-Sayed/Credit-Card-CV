```python
import cv2
import numpy as np

def nothing(x):
    pass

image = cv2.imread("E:/vs code projects/CreditCardProject/TC/05.jpg")  # Load in grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# canny=cv2.Canny(gray, 50, 150)

# Create a window
cv2.namedWindow('Morphology Operations')

# Create trackbars for kernel size and operation type
cv2.createTrackbar('threshold', 'Morphology Operations', 0, 255, nothing)

while True:
    # Get the current positions of the trackbars
    ksize = cv2.getTrackbarPos('threshold', 'Morphology Operations')
    processed_image = cv2.threshold(gray, ksize, 255, cv2.THRESH_BINARY_INV)[1]

    # Show the processed image
    cv2.imshow('Morphology Operations', processed_image)

    # Break the loop if the user presses the 'Esc' key
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
