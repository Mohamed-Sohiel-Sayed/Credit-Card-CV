```python
import cv2 
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import utils as ut
import refFont

font = refFont.Font()
```

```python
# Load the image
image_path = './TC/14.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Step 1: Contrast enhancement using Histogram Equalization
equalized_image = cv2.equalizeHist(image)

# Step 2: Adaptive Thresholding
thresholded_image = cv2.adaptiveThreshold(equalized_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 11, 2)

# Step 3: Edge Detection using Canny
edges = cv2.Canny(thresholded_image, 100, 200)

# Display the results
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(2, 2, 2)
plt.title('Histogram Equalized')
plt.imshow(equalized_image, cmap='gray')
plt.subplot(2, 2, 3)
plt.title('Adaptive Thresholding')
plt.imshow(thresholded_image, cmap='gray')
plt.subplot(2, 2, 4)
plt.title('Edges Detected')
plt.imshow(edges, cmap='gray')
plt.tight_layout()
plt.show()
```
